# Video Access

```python
video=cv2.VideoCapture('Champion (Official Song).mp4')
while video.isOpened():
    ret,frame=video.read()
    if ret:
        video1=cv2.resize(frame,(600,400))
        cv2.imshow("video player",video1)
        
        if cv2.waitKey(25) & 0xff==ord('q'):
            break
    else:
        break
video.release()
cv2.destroyAllWindows()
```

# Different Different View in a grid

```python
camera_view=cv2.VideoCapture(0)
video1=cv2.VideoCapture('Champion (Official Song).mp4')
video2=cv2.VideoCapture('Waka Waka.mp4')
video3=cv2.VideoCapture('O Antava.mp4')
while camera_view.isOpened():
    ret1,frame=camera_view.read()
    ret2,frame2=video1.read()
    ret3,frame3=video2.read()
    if ret1:
        frame=cv2.resize(frame,(600,300))
        frame2=cv2.resize(frame2,(600,300))
        frame3=cv2.resize(frame3,(600,300))
        image=cv2.imread('noidea.jpg')
        image=cv2.resize(image,(600,300))
        
        frame_2=np.hstack((frame,frame2))
        frame_02=np.hstack((frame3,image))
        frame_4=np.vstack((frame_2,frame_02))
        cv2.imshow("video player",frame_4)
        
        if cv2.waitKey(25) & 0xff==ord('q'):
            break
    else:
        break
camera_view.release()
cv2.destroyAllWindows()
```
