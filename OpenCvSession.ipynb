# Video Access

```python
Video=cv2.VideoCapture('Champion (Official Song).mp4')
while Video.isOpened():
    ret,frame=Video.read()
    if ret:
        video=cv2.resize(frame,(600,400))
        cv2.imshow("Random Player",video)
        
        if cv2.waitKey(10) & 0xff==ord('q'):
            break
    else:
        break
Video.release()
cv2.destroyAllWindows()
```

# Different Data sources for different grid

```python
import numpy as np
import cv2
camera_view=cv2.VideoCapture(0)
video=cv2.VideoCapture('Champion (Official Song).mp4')
video2=cv2.VideoCapture('O Antava.mp4')
while camera_view.isOpened():
    ret1,frame1=camera_view.read()
    ret2,frame2=video.read()
    ret3,frame3=video2.read()
    if ret1:
        video=cv2.resize(frame1,(300,150))
        video1=cv2.resize(frame2,(300,150))
        video2=cv2.resize(frame3,(300,150))
        image=cv2.imread('noidea.jpg')
        image=cv2.resize(image,(300,150))
        
        video_1=np.hstack((video,video1))
        video_2=np.hstack((video2,image))
        video_4=np.vstack((video_1,video_2))
        cv2.imshow("Random Player",video_4)
        
        if cv2.waitKey(10) & 0xff==ord('q'):
            break
    else:
        break
video.release()
cv2.destroyAllWindows()
```

```python
import os
import cv2
import numpy as np
```

```python
os.mkdir("video_to_img")
video=cv2.VideoCapture('O Antava.mp4')
img_count=1
while video.isOpened():
    ret,frame=video.read()
    
    if not ret:
        print("video does not have any of frame")
        break
    is_img_write=cv2.imwrite(f"video_to_img\img{img_count}.jpeg",frame)
    
    if is_img_write:
        print(f"image save at video_to_img\img{img_count}.jpeg")
    cv2.imshow("video",frame)
    cv2.waitKey(10) & 0xff==ord('q')
    img_count +=1
video.release()
cv2.destroyAllWindows()
    
```
